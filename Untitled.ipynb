```python
import os
import glob

import folium
import gpxpy
import tcxparser

import numpy as np
import pandas as pd
from geopy.geocoders import Nominatim
```

```python
geolocator = Nominatim()
location = geolocator.geocode("Montreal Quebec")
lat_check = float(location.raw['lat'])
lon_check = float(location.raw['lon'])

gpxdata = glob.glob('gpxdata/*.gpx')
tcxdata = glob.glob('tcxdata/*.tcx')
```

```python
filename = gpxdata[0][8:-4]
```

```python
gpx_file = open('gpxdata/'+filename+'.gpx', 'r')
gpx = gpxpy.parse(gpx_file)
```

```python
print(gpx)
```

```python
tcx_file = open('tcxdata/'+filename+'.tcx', 'r')
tcx = tcxparser.TCXParser(tcx_file)
```

```python
print(tcx.pace)
tcx.speed
```

```python
lat = []
lon = []
```

```python
for track in gpx.tracks:
    for segment in track.segments:
        for point in segment.points:
            lat.append(point.latitude)
            lon.append(point.longitude)

points = zip(lat,lon)
points = [item for item in zip(lat,lon)]
```

```python
import matplotlib.cm as cm
import matplotlib
```

```python
def get_color(hr):
    norm = matplotlib.colors.Normalize(vmin=40, vmax=200)
    cmap = cm.nipy_spectral
    m = cm.ScalarMappable(norm=norm, cmap=cmap)
    return m.to_rgba(hr)[:-1]
```

```python
for hr in tcx.hr_values():
    print(get_color(hr))
```
